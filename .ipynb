```python
from selenium.webdriver import Chrome
from time import sleep
driver = Chrome()
url ='https://www.linkedin.com/jobs/'
driver.get(url)

#Clicking sign-in page
element_signin = driver.find_element_by_xpath("//a[contains(@class, 'nav__button-secondary')]")
element_signin.click()
sleep(1)

#Sign-in
element_username = driver.find_element_by_id('username')
element_username.send_keys('hykimusa@gmail.com')
element_pw = driver.find_element_by_id('password')
element_pw.send_keys('Linkedinpw4$')
sign_in_button = driver.find_element_by_xpath('//button[contains(@class, "btn__primary--large from__button--floating mercado-button--primary")]')
sign_in_button.click()
sleep(3)

# Searching for a job of interest
# Creating a function
def find_job_on_linkedin(job_title, city, province_state):
    element_job_title = driver.find_element_by_id('jobs-search-box-keyword-id-ember49')
    element_job_title.click()
    element_job_title.clear()
    element_job_title.send_keys(job_title)
       
    element_loc = driver.find_element_by_id('jobs-search-box-location-id-ember49')
    element_loc.click()
    element_loc.clear()
    element_loc.send_keys(f'{city}, {province_state}')
    sleep(3)
    
    element_search = driver.find_element_by_xpath("//button[text()='Search']")
    element_search.click()
    
# Test the function
find_job_on_linkedin('scientist', 'Toronto', 'Ontario')
```
